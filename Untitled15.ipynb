```python
import pandas as pd
import random
# Generate enhanced synthetic traffic accident data
locations = ["New york", "Los Angeles", "Chicago", "Houston", "Miami"]
weather_conditions = ["Clear", "Rain", "Fog", "Snow", "Storm"]
accident_reasons = ["Speeding", "Drunk Driving", "Red Light Violation", "Reckless Driving", "Distracted Driving"]
times_of_day = ["Morning", "Afternoon", "Evening", "Night"]
accident_types = ["Head-On Collision", "Rear-End Collision", "Side Impact", "Pedestrian Involved"]
road_conditions = ["Dry", "Wet", "Icy", "Potholes"]
enhanced_data = []
for i in range(1, 201):  # Generating 200 accident records
    entry = {
        "Accident_ID": i,
        "Location": random.choice(locations),
        "Date": pd.date_range(start="2024-01-01", periods=200, freq="D")[i-1],
        "Time_of_Day": random.choice(times_of_day),
        "Severity": random.randint(1, 5),
        "Weather": random.choice(weather_conditions),
        "Road_Condition": random.choice(road_conditions),
        "Accident_Type": random.choice(accident_types),
        "Speed_Limit": random.choice([30, 40, 50, 60, 70, 80]),
        "Reason": random.choice(accident_reasons),
        "Vehicles_Involved": random.randint(1, 5),
        "Injuries": random.randint(0, 10),
        "Fatalities": random.randint(0, 3)  # New fatality count parameter
    }
    enhanced_data.append(entry)
# Create DataFrame
df_enhanced_accidents = pd.DataFrame(enhanced_data)
# Save to CSV
df_enhanced_accidents.to_csv("enhanced_traffic_accidents.csv", index=False)

# Display first few rows
df_enhanced_accidents.head()




```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Date' column to datetime format
df_enhanced_accidents["Date"] = pd.to_datetime(df_enhanced_accidents["Date"])

# Set plot style
sns.set_style("whitegrid")

# Count of accidents by severity
plt.figure(figsize=(8,5))
sns.countplot(x="Severity", data=df_enhanced_accidents, hue="Severity", palette="Reds", legend=False)
plt.xlabel("Accident Severity Level")
plt.ylabel("Number of Accidents")
plt.title("Distribution of Accident Severity Levels")
plt.show()

```

```python
# Accidents by weather conditions
plt.figure(figsize=(10,5))
sns.countplot(y="Weather", data=df_enhanced_accidents, palette="coolwarm", hue="Weather", order=df_enhanced_accidents["Weather"].value_counts().index, legend=False)
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.title("Accidents Under Different Weather Conditions")
plt.show()

```

```python

# Fatalities by accident type
plt.figure(figsize=(10,5))
sns.barplot(x="Accident_Type", y="Fatalities", data=df_enhanced_accidents,hue="Accident_Type", estimator=sum, errorbar=None, palette="magma", legend=False)
plt.xlabel("Accident Type")
plt.ylabel("Total Fatalities")
plt.title("Total Fatalities by Accident Type")
plt.xticks(rotation=45)
plt.show()

```

```python
# Accidents over time
plt.figure(figsize=(12,5))
df_enhanced_accidents.groupby("Date").size().plot(kind="line", marker="o", color="blue", linewidth=2)
plt.xlabel("Date")
plt.ylabel("Number of Accidents")
plt.title("Accident Trends Over Time")
plt.grid()
plt.show()
```
